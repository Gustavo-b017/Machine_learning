```python
import math
import matplotlib.pyplot as plt

def pitagoras(a, b):
  return math.sqrt(a**2 + b**2)

def distancia_matrix(x, y):
  matrix = []

  for i in range(len(x)):
    lista = []  
    for j in range(len(x)):
      dist = pitagoras(x[i], y[j])
      d = round(dist, 2)
      lista.append(d)
  return lista

x= [1,2,3,4,5,6]
y = [3,1,4,7,2,5]

matrix = distancia_matrix(x, y)

for linha in matrix:
  print(linha)

plt.plot(x, y, 'ro')
plt.show()
```
